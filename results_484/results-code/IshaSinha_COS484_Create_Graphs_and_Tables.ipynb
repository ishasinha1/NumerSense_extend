```python
import matplotlib.pyplot as plt
```

```python
def draw_graph(x, height, colors=None):
  fig, ax = plt.subplots()

  if colors is not None:
    ax.bar(x, height, color=colors, linewidth=1, edgecolor='black')
  else:
    ax.bar(x, height, linewidth=1, edgecolor='black')

  ax.set_ylabel('# of Examples')
  ax.set_xlabel('Word')

  plt.show()
```

```python
# True distribution of size
size = ['larger', 'smaller']
size_col = ['red', 'blue']
draw_graph(size, [1760, 1368], size_col)
```

```python
# Distribution of BERT predictions for size
draw_graph(size, [2657, 471], size_col)
```

```python
# Distribution of RoBERTa predictions for size
draw_graph(size, [2760, 368], size_col)
```
